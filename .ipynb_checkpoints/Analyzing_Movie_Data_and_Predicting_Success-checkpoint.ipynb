# 3. Data Cleaning and Feature Engineering

Once the data was mined, we checked the data and found out that things were not as expected. There were many instances of missing data. In many cases, the format of the data was confusing and not one which we could feed our training model or use for visualizations. After carefully studying the mined data and our requirements, we came up with the following specifications for cleaning the data and generating useful features out of it:

- IMDb returns a movie object for every movieID. It had only cast IDs and director IDs. Using these, fetch corresponding cast names and director names from IMDb.
- Consider only 'movies' not 'TV series'.
- The movie object returned also did not have the budget and revenue data. 'Update' the object using IMDbPy to fetch the budget, revenue and release month.
- Runtime was in the string format with the country and varying characters around the actual runtime. 
  Eg: USA:96, Switzerland:128::(Locarno Film Festival)
  Extract the number from that and store as float.
- Budget, revenue were also strings with prefixes of the currency before the actual value. Extract the number from that and store as float. 
- Opening weekend was a complex object with details of opening revenue in it along with the date. The date was also like '19 September 1983'. Extract the month out of it.
- There were many duplicates. Drop movies with the movieID already covered.
- Create one-hot encoding with genres. There were a total of 23 different genres. The one hot encoding would be fed into the model.

The following code implements the above specifications

```python
#CLeaning the data as needed from the raw data recieved from IMDb 
import re
def clean_movie(m):
    d={}
    if m and m['kind'] and m['kind']=='movie':
        ia.update(m,'business')
        if m.get('cast'):
            cast = m.get('cast')
            d['cast0']= cast[0].personID
            d['cast0_name'] = cast[0]['name'].encode('utf-8')
            if(len(cast)>1):
                d['cast1']= cast[1].personID
                d['cast1_name'] = cast[1]['name'].encode('utf-8')
            if(len(cast)>2):
                d['cast2']= cast[2].personID
                d['cast2_name'] = cast[2]['name'].encode('utf-8')
        if m.get('director'):
            drctr=m.get('director')[0]
            d['director']=drctr.personID
            d['director_name'] = drctr['name'].encode('utf-8')
        
        d['year']=m.get('year')
        d['rating']=m.get('rating')
        if m.get('runtimes'):
            srch=re.search(r'[^0-9]*(\d+)', str(m.get('runtimes')[0].encode('utf-8')))
            if srch and srch.groups():
                d['runtime'] = float(srch.groups()[0])
            else:
                d['runtime'] = float(m.get('runtimes')[0])
            
        
        d['top 250 rank']= m.get('top 250 rank')
        d['mpaa']= extract_mpaa(m.get('mpaa'))

        if m.get('languages'):
            d['language']=m.get('languages')[0]

        genres_main = ['Action','Animation','Comedy','Documentary','Family','Film-Noir','Horror','Musical','Romance','Sport','War','Adventure','Biography','Crime','Drama','Fantasy','History','Music','Mystery','Sci-Fi','Thriller','Western','Short']    
        for gen in genres_main:
            
            d['genre_'+gen] = 0
           
        genres=m.get('genres')
        if(genres):
            for g in genres:
                if(g in genres_main):
                    d['genre_'+g]= 1
        
        if m.get('business'):
            bis=m.get('business')
            if 'gross' in bis:
                rev=bis.get('gross')[0]
                #print rev
                temp= re.search(r'[^0-9]*([0-9,.]+)', rev).groups()[0]
                d['revenue']= long(re.sub("[^\d]", "", temp))

            if 'budget' in bis:
                bud=bis.get('budget')[0]
                #print bud
                temp= re.search(r'[^0-9]*([0-9,.]+)', bud).groups()[0]
                d['budget']=long(re.sub("[^\d]", "", temp))

            if 'opening weekend' in bis:
                opstr=bis.get('opening weekend')[0]
                d['month']=extract_month(opstr)
                
            d['movie_id'] = m.movieID
    
            d['votes']= m.get('votes')
    return d

def extract_mpaa(strng):
    if(strng):
        vals = strng.split(' ')
        return vals[1]
    else:
        return None

def extract_month(strng):
    months = ['January','February','March','April','May','June','July','August','September','October','November','December']
    for month in months:
        if month in strng:
            return month
    return None
         
```

## 4.2. Average movie length per language

In our dataset, we have movies in a lot of languages! We were curious to know the runtime of the movies as per the language in which it was created. So we took the top 20 languages in which movies are made and plotted the average runtime of movies in those languages. On the X axis, the languages are present in decreasing order of popularity (in this case, the number of movies made in them)

```python
from collections import Counter
def process_runtime(rntime):
    
    srch=re.search(r'[^0-9]*(\d+)', str(rntime))
    if srch and srch.groups():
        return float(srch.groups()[0])
    else:
        return float(rntime)
```

```python
    moviesdf['runtime'] = moviesdf['runtime'].apply(process_runtime)

languages_to_print = []


moviesdf = moviesdf[moviesdf['runtime']>0.0]


movieslang = moviesdf[pd.notnull(moviesdf['language'])]

languages_set = Counter(movieslang['language']).most_common(20)
for ele in languages_set:
    languages_to_print.append(ele[0])

languages_values = []

movlang = movieslang.groupby(['language'])

mean_runtime = movlang['runtime'].mean()

for lang in languages_to_print:
    
    
        languages_values.append(mean_runtime.loc[lang])


plt.bar(range(len(languages_to_print)), languages_values)
plt.xticks(range(len(languages_to_print)), languages_to_print)
plt.show()
```

This graph turned out to be as expected. Movies in Indian languages are usually longer due to the fact that they have lot of songs and dances. Rest of the languages have more or less the same runtimes. The Scandanavian ones - Finnish and Danish have relatively shorter movies.

# 5. Predicting missing Revenue data

Before we could model the data for our primary focus, we ran into an unexpected hurddle. We realized that almost half of our movie records did not have revenue data. IMDb does not have revenue data for all the movies. This was something we did not expect. Hence, we decided to use predictive modeling to predict the missing revenue data. Here, we would use the movies with revenue data to train and evaluate which model gives the best results. But before passing data into the model, some pre-processing was required.

## 5.1. Pre-processing of data

Before passing data into the sci-kit learn models, we had to do some pre-processing. 

- Non numeric features irrelevant to the revenue like cast names and director names were dropped, only their IDs were retained.
- Other categorical features relevant to predict revenue like the language of the movie, certificate, month of release were converted to numeric data by allocating numbers to each category. 
- Movie_id was dropped since it is unique for every movie and has nothing to do with the revenue. Top 250 rank of the movie was also dropped since it was a very sparse feature.
- Empty values for the remaining features were filled with the mean of the feature (in case of nunmeric features) or with the mode of the feature (in case of a categorical feature)

The following code pre-processes the data as per the above specifications and creates a train test split of 85% and 15% for evaluating the models.

```python
#Creating Movies DF for features with non-NA revenue values
moviesdfnr = pd.read_csv("final_movie_dataset.csv")
moviesdfnr = moviesdfnr.drop_duplicates('movie_id')

moviesdfnr = moviesdfnr[moviesdfnr['rating']>0.0]
moviesdfnr.drop(['director_name','cast0_name','cast1_name','cast2_name','votes'],1,inplace=True)

moviesdfnr['budget'].fillna(moviesdfnr['budget'].median(skipna=True),inplace=True)
moviesdfnr['runtime'] = moviesdfnr['runtime'].apply(process_runtime)
moviesdfnr['runtime'].fillna(moviesdfnr['runtime'].median(skipna=True),inplace=True)
moviesdfnr['month'].fillna(moviesdfnr['month'].mode()[0],inplace=True)
moviesdfnr['mpaa'].fillna(moviesdfnr['mpaa'].mode()[0],inplace=True)
moviesdfnr['cast2'].fillna(0,inplace=True)
moviesdfnr['cast1'].fillna(0,inplace=True)
moviesdfnr['cast0'].fillna(0,inplace=True)
moviesdfnr['director'].fillna(0,inplace=True)
moviesdfnr['language'].fillna(moviesdfnr['language'].mode()[0],inplace=True)

#Preparing data for use with sklearn models
movieids=moviesdfnr['movie_id']
revdf=moviesdfnr.drop(['movie_id','top 250 rank'],1)

for col in ['language','mpaa','month']:
    vals=list(revdf[col].unique())
    revdf[col]=revdf[col].apply(vals.index)

revdf=revdf.rename(columns = {'revenue':'label'})

revtrdf = revdf[revdf['label']>0]
revtrlabels=[int(x) for x in revtrdf['label']]

revtedf = revdf[revdf['label'].apply(np.isnan)]
temovies=movieids[revdf['label'].apply(np.isnan)].reset_index()
dfids=np.array(temovies['index'])
del temovies['index']
temovies=np.array(temovies).flatten()

revtrdf=revtrdf.drop('label',1)
revtedf=revtedf.drop('label',1)

rev_train, rev_test, rlabel_train, rlabel_test = train_test_split(revtrdf, revtrlabels, test_size=0.15, random_state=0)
```

## 5.2. Base case: Mean predictor

To use as the base case for our comparisions, we decided to see the performance of a simple mean predictor that is we assumed a model which would predict the mean revenue of existing data for all the test data. The error in all the models is:

Error = Actual revenue - Predicted revenue

```python
#Revenue prediction with median
rev_pred=[np.median(rlabel_train)]*len(rev_test)
print "Mean error is:",np.mean(map(sub,rev_pred,rlabel_test))
```

```python
#Revenue predicton with RF
clf = RandomForestRegressor(n_estimators=50)
clf.fit(rev_train, rlabel_train)
rev_pred = clf.predict(rev_test)
print "Mean error is:",np.mean(map(sub,rev_pred,rlabel_test))
```

## 5.4. Logistic Regression based predictor

We then decided to use Logistic regression with normalized input. Here is the performance:

```python
#Revenue predicton with Logistic Regression and normalized input
clf=LogisticRegression()
min_max_scaler = preprocessing.MinMaxScaler()
rev_train_norm = min_max_scaler.fit_transform(rev_train)
rev_test_norm = min_max_scaler.fit_transform(rev_test)
clf.fit(rev_train_norm, rlabel_train)
rev_pred = clf.predict(rev_test_norm)
print "Mean error is:",np.mean(map(sub,rev_pred,rlabel_test))
```

## 5.5.  Filling in missing Revenue values

The performance of logistic regression was the best. Forests performed an order of magnitude worse than the base case mean predictor. This is because decision trees are usually good classifiers but not good predictors. Also, they are known not to perform well with data having high variance, which was there in our case. Even logistic regression, we tried parameter tuning with C values of 1000, 10000 and 1000000 but did not see much difference in the error rate.

Hence, we used logistic regression on all the records which had revenue (our training data) and then used that model to fill the missing revenue values. Once the revenue values were predicted, they were put into the main dataframe using their movie ids.

```python
#Filling in Nan revenue values
clf=LogisticRegression()
clf.fit(revtrdf, revtrlabels)
revna_pred = clf.predict(revtedf)
```

## 6.1 Naive Bayes with Gaussians and Categorical Predictors

We first attempted modeling the data using maximum likelihood estimation. We re-used the code that we wrote for the course homework for this classifier. The continuous data from the 'year','budget','revenue','runtime' and 'top 250 rank' columns were modelled using Gaussian predictors, and we used categorical predictor for the non continuous data in all other columns. We used this model to classify the rating into 10 possible ratings, ranging from 0-9.

```python
#Generating the movies features from the saved dataset
moviesdf = pd.read_csv("final_movie_dataset.csv")
revendf=pd.read_csv("predicted_revenues.csv")
revendf.set_index('df_id',inplace=True)
moviesdf = moviesdf.drop_duplicates('movie_id')

moviesdf = moviesdf[moviesdf['rating']>0.0]
moviesdf['rating']=moviesdf['rating'].apply(int).subtract(1)
moviesdf=moviesdf.rename(columns = {'rating':'label'})
moviesdf.drop(['director_name','cast0_name','cast1_name','cast2_name','votes'],1,inplace=True)

#Filling missing continuous values with their median
moviesdf['budget'].fillna(moviesdf['budget'].median(skipna=True),inplace=True)
moviesdf['runtime'] = moviesdf['runtime'].apply(process_runtime)
moviesdf['runtime'].fillna(moviesdf['runtime'].median(skipna=True),inplace=True)

#Filling missing categorical values with their mode
moviesdf['month'].fillna(moviesdf['month'].mode()[0],inplace=True)
moviesdf['mpaa'].fillna(moviesdf['mpaa'].mode()[0],inplace=True)
moviesdf['language'].fillna(moviesdf['language'].mode()[0],inplace=True)

moviesdf['cast2'].fillna(0,inplace=True)
moviesdf['cast1'].fillna(0,inplace=True)
moviesdf['cast0'].fillna(0,inplace=True)
moviesdf['director'].fillna(0,inplace=True)

moviesdf.loc[revendf.index,'revenue']=revendf['revenue']
```

```python
"""Re-used from course assignment"""

class GaussianPredictor:
    def __init__(self, x, y) :
        self.k = len(y.unique())
        self.mu = np.zeros(self.k)
        self.sigma = np.zeros(self.k)
        vals=y.unique()
        for i in vals:
            indices=np.where(y==i)[0]
            xi=x[indices]
            self.mu[i]=np.mean(xi)
            self.sigma[i]=np.std(xi)
    
    def partial_log_likelihood(self, x):
        musig=zip(self.mu,self.sigma)
        return np.log([[norm.pdf(i, j[0], j[1]) for j in musig] for i in x])
        
        pass
```

```python
"""Re-used from course assignment"""

class CategoricalPredictor:
    def __init__(self, x, y, alpha=1) :
        self.p = {}
        classes=y.unique()
        l=len(classes)
        self.k = len(classes)
        vals=y.unique()
        featurevals=x.unique()
        #print len(featurevals)
        lf=len(featurevals)
        for i in classes:
            indices=np.where(y==i)[0]
            xi=x[indices]
            #print x,xi
            for f in featurevals:
                if f not in self.p:
                    self.p[f]=np.zeros(l)
                fcount=list(xi).count(f)
                self.p[f][i]=(fcount+alpha)/(float(len(xi) +(lf*alpha)))
                    
    def partial_log_likelihood(self, x):
        return np.log([self.p[i] for i in x])
        pass
```

```python
"""Re-used from course assignment"""

class NaiveBayesClassifier:
    def __init__(self, df, alpha=1):
        self.predictor={}
        classes=df['label'].unique()
        label=list(df['label'])
        l=len(label)
        k=len(classes)
        self.log_prior=np.array(np.log([(label.count(i)+alpha)/float(l+(alpha*k)) for i in range(0,k)]))
        
        gauscols= ['year','budget','revenue','runtime','top 250 rank']
        catcols=list(set(df.columns)-set(gauscols))
        for col in gauscols:
            if col in df.columns:
                #print col
                self.predictor[col]=GaussianPredictor(df[col], df['label'])
        for col in catcols:
            #print col
            if col!='index' and col!='label' and col!='movie_id':
                self.predictor[col]=CategoricalPredictor(df[col], df['label'],alpha)
        pass

    def predict(self, x):
        predictions=[self.predictor[pred].partial_log_likelihood(x[pred]) for pred in self.predictor]
        matsum=0
        for i in range(0,len(predictions)):
            if not i:
                matsum=predictions[i]
            else:
                matsum=map(self.matsumfunc, matsum, predictions[i])
        matsum2=[map(add,i,self.log_prior) for i in matsum]
        return [np.argmax(z) for z in matsum2]
        pass
    
    def matsumfunc(self,a,b):
        return map(add,a,b)
```

```python
"""Re-used from course assignment"""

#Evaluates the performance of the model on the basis of error rate
def evaluate(y_hat, y):
    return np.mean(y_hat!=y)
    pass
```

## 6.5 Hyper-parameters selection using Grid Search

We ran a grid search over the set of possible values for each of the following hyper-parameters for Random Forests: Maximum depth, number of features to be considered for a split, minimum samples needed for a split and needed to be present at a leaf node, with/without bootstrap samples for tree-building, and the function used to measure the quality of splits.

```python
#Function to report the performance of each combination in the grid
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```python
#RF gridSearch
clf = RandomForestClassifier(n_estimators=50)
param_grid = {"max_depth": [3, None],
              "max_features": [0.2, 0.5, 0.8, 1],
              "min_samples_split": [2, 6, 10],
              "min_samples_leaf": [1, 2, 6, 10],
              "bootstrap": [True, False],
              "criterion": ["gini", "entropy"]}

# run grid search
grid_search = GridSearchCV(clf, param_grid=param_grid)
start = time()
grid_search.fit(X_train10, y_train10)

print("GridSearchCV took %.2f seconds for %d candidate parameter settings."
      % (time() - start, len(grid_search.cv_results_['params'])))
report(grid_search.cv_results_)

"""
References:
1) http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
2) http://scikit-learn.org/stable/auto_examples/plot_compare_reduction.html
"""
```

The grid search returned the best set of hyper parameters, with the values as shown above.
